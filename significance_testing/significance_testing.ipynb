# Read data
Read dataframe with model performance metrics

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/classification reports - final results.csv', header=2
                 ).rename(columns={
    'macro F1': 'dev cr-bal macro F1',
    'one vs all F1': 'dev cr-bal one vs all F1',
    'ROC AUC (where applicable)': 'dev cr-bal ROC AUC',
    'Accuracy': 'dev cr-bal Accuracy',

    'macro F1.1': 'dev tr macro F1',
    'one vs all F1.1': 'dev tr one vs all F1',
    'ROC AUC (where applicable).1': 'dev tr ROC AUC',
    'Accuracy.1': 'dev tr Accuracy',

    'macro F1.2': 'test cr-bal macro F1',
    'one vs all F1.2': 'test cr-bal one vs all F1',
    'ROC AUC (where applicable).2': 'test cr-bal ROC AUC',
    'Accuracy.2': 'test cr-bal Accuracy',

    'macro F1.3': 'test tr macro F1',
    'one vs all F1.3': 'test tr one vs all F1',
    'ROC AUC (where applicable).3': 'test tr ROC AUC',
    'Accuracy.3': 'test tr Accuracy',
})
df.dropna(subset=['model'], inplace=True)
df['model_name_full'] = df['model'] + '_' + df['balanced / full data']
df
```

```python
for col in [
    'dev cr-bal one vs all F1',
    'dev tr one vs all F1',
    'test cr-bal one vs all F1',
    'test tr one vs all F1']:
  df[f'{col}_1'] = df['test cr-bal one vs all F1'].apply(lambda x: x.split('\n')[0].split(': ')[1]).astype(float)
  df[f'{col}_2'] = df['test cr-bal one vs all F1'].apply(lambda x: x.split('\n')[1].split(': ')[1]).astype(float)
  df[f'{col}_3'] = df['test cr-bal one vs all F1'].apply(lambda x: x.split('\n')[2].split(': ')[1]).astype(float)

df
```

# ASO: Macro F1

Format data as input to `multi_aso`. We want to rank the models based on their macro F1 scores on the test data set

```python
test_macro_f1_scores = {}
models = df['model_name_full'].unique()
for m in models:
  scores = df[df['model_name_full'] == m]['test cr-bal macro F1'].to_numpy()
  test_macro_f1_scores[m] = scores
test_macro_f1_scores
```

Run ASO test

How to read (based on the `deepsig` documentation):

each cell shows the ASO score for the models in the given row and column. If the score is **smaller** than the threshold t (default `t = 0.5`), the **row** model is **better** than the **column** model.

```python
aso_df = multi_aso(test_macro_f1_scores, confidence_level=0.95, return_df=True, seed=42)
aso_df
```

```python
aso_df.reset_index(names=['model'], inplace=True)
aso_df.to_csv('pairwise_aso_tables/pairwise_aso_macro_F1.csv', index=False)
```

```python
t = 0.4
better_than, worse_than = {}, {}

for m_this in models:
  this_aso_scores = aso_df[aso_df['model'] == m_this].iloc[0].to_dict()
  better_than[m_this] = [m for m in models if m != m_this and this_aso_scores[m] < t]
  worse_than[m_this] = [m for m in models if m != m_this and this_aso_scores[m] > t]

better_than

eq_classes = {m: {'models': [m], 'rank': 100} for m in models}
for m_this in models:
  for m in better_than[m_this]:  # if m is both better and worse, they're equal
    if m_this in better_than[m]:
      eq_classes[m_this]['models'].append(m)
      eq_classes[m_this]['rank'] = min(eq_classes[m_this]['rank'],
                                       len(worse_than[m]))
  for m in worse_than[m_this]:
    if m_this in worse_than[m]:
      eq_classes[m_this]['models'].append(m)
      eq_classes[m_this]['rank'] = min(eq_classes[m_this]['rank'],
                                       len(worse_than[m]))

ranking = {}
for m in eq_classes:
  r = eq_classes[m]['rank']
  if r in ranking:
    ranking[r].append(m)
  else:
    ranking[r] = [m]

ranking = dict(sorted(ranking.items()))
ranking
```

```python
better_than
```

```python
t = 0.4

for m_this in models:
  this_aso_scores = aso_df[aso_df['model'] == m_this].iloc[0].to_dict()
  better_than, worse_than = [], []

  for m_other in [m for m in models if m != m_this]:
    if this_aso_scores[m_other] < t:
      better_than.append(m_other)
    else:
      worse_than.append(m_other)

  rank = len(worse_than) + 1
  avr_macro_F1 = test_macro_f1_scores[m_this].mean()
  print(f'{m_this}: {rank}')
  print(f'Average macro F1: {avr_macro_F1:.3}')
  print(f'Better than: {", ".join(better_than)}')
  print(f'Worse than: {", ".join(worse_than)}')
  print()
```
