```python
from deepsig import aso, multi_aso
```

```python
import os
os.makedirs('pairwise_aso_tables', exist_ok=True)
```

# Read data
Read dataframe with model performance metrics

```python
import pandas as pd
```

```python
df = pd.read_csv('project_results - all metrics for all models and seeds.csv', header=1
                 ).rename(columns={
    'macro F1': 'dev macro F1',
    'macro F1.1': 'test macro F1',
    'one vs all F1': 'dev one vs all F1',
    'one vs all F1.1': 'test one vs all F1',
    'ROC AUC (where applicable)': 'dev ROC AUC',
    'ROC AUC (where applicable).1': 'test ROC AUC',
    'Accuracy': 'dev Accuracy',
    'Accuracy.1': 'test Accuracy'
})
df.dropna(subset=['model'], inplace=True)
df['model_name_full'] = df['model'] + '_' + df['balanced / unbalanced data']
df
```

# ASO: Macro F1

Format data as input to `multi_aso`. We want to rank the models based on their macro F1 scores on the test data set

```python
test_macro_f1_scores = {}
models = df['model_name_full'].unique()
for m in models:
  scores = df[df['model_name_full'] == m]['test macro F1'].to_numpy()
  test_macro_f1_scores[m] = scores
test_macro_f1_scores
```

Run ASO test

How to read (based on the `deepsig` documentation):

each cell shows the ASO score for the models in the given row and column. If the score is **smaller** than `t = 0.5`, the **row** model is **better** than the **column** model.

```python
aso_df = multi_aso(test_macro_f1_scores, confidence_level=0.95, return_df=True, seed=42)
aso_df
```

```python
aso_df.reset_index(names=['model'], inplace=True)
aso_df.to_csv('pairwise_aso_tables/pairwise_aso_macro_F1.csv', index=False)
```

```python
t = 0.4

for m_this in models:
  this_aso_scores = aso_df[aso_df['model'] == m_this].iloc[0].to_dict()
  better_than, worse_than = [], []

  for m_other in [m for m in models if m != m_this]:
    if this_aso_scores[m_other] < t:
      better_than.append(m_other)
    else:
      worse_than.append(m_other)

  rank = len(worse_than) + 1
  avr_macro_F1 = test_macro_f1_scores[m_this].mean()
  print(f'{m_this}: {rank}')
  print(f'Average macro F1: {avr_macro_F1:.3}')
  print(f'Better than: {", ".join(better_than)}')
  print(f'Worse than: {", ".join(worse_than)}')
  print()
```
